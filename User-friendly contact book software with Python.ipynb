```python
class Contact:
    def __init__(self, id, name, surname, phone, email):
        """
        Definizione nome, cognome, telefono, mail
        """
        self.id = id
        self.name = name
        self.surname = surname
        self.phone = phone
        self.email = email

    def __str__(self):
      info = f"\n{self.id} Name: {self.name}, Surname: {self.surname}, Phone number: {self.phone}, Email: {self.email}"
      return info


    def to_dict(self):
      return self.__dict__
```

### Add a new contact

```python
# POINT 1 - ADD A NEW CONTACT

chosen_option = None
new_id = None
new_name = None
new_surname = None
new_phone = None
new_email = None
contacts = {}

i = 1

def add_new_contact(contacts, chosen_option, i, new_id, new_name, new_surname, new_phone, new_email):

      if chosen_option == 1:
          print("\nYou have chosen to add a new contact!\n")
          new_id = i

          while new_name is None:
              try:
                  new_name = input(f"Please, insert below the NAME of the new contact"+f"\n{exclamation_mark} Take note: it must contain letters!\n \n")
                  if new_name.isnumeric():
                      new_name = None
                      raise ValueError(f"\n{cross_error} Oops! Something went wrong! {cross_error}"+"\nPlease, insert a correct name! \n")
              except ValueError as e:
                  print(e)

          while new_surname is None:
              try:
                  new_surname = input("\nPlease, insert here the SURNAME of the new contact: "+f"\n{exclamation_mark} Take note: it must contain letters!\n \n")
                  if new_surname.isnumeric():
                      new_surname = None
                      raise ValueError(f"\n{cross_error} Oops! Something went wrong! {cross_error}"+"\nPlease, insert a correct surname! \n")
              except ValueError as e:
                  print(e)

          while new_phone is None:
              try:
                  new_phone = input("\nPlease, insert here the PHONE NUMBER of the new contact: "+f"\n{exclamation_mark} Take note: it must contain numbers!\n \n")
                  if not new_phone.isnumeric():
                      new_phone = None
                      raise ValueError(f"\n{cross_error} Oops! Something went wrong! {cross_error}"+"\nPlease, insert a correct phone number! \n")
                  new_phone = str(new_phone) #if the number starts with 0, it is better to save it as a str to avoid losing it
              except ValueError as e:
                  print(e)

          while new_email is None:
              try:
                  new_email = input("\nPlease, insert here the EMAIL ADDRESS of the new contact: "+f"\n{exclamation_mark} Take note: it must contain a"+" @ "+"character!\n \n")
                  if (new_email.isnumeric()) or ("@" not in new_email):
                      new_email = None
                      raise ValueError(f"\n{cross_error }Oops! Something went wrong! {cross_error}"+"\nPlease, insert a correct email address! \n")
              except ValueError as e:
                  print(e)
          new_contact = Contact(id=new_id, name=new_name, surname=new_surname, phone=new_phone, email=new_email)
          contacts[f"{new_contact.id} {new_contact.name} {new_contact.surname}"] = [new_contact.phone, new_contact.email]

          print("\nHere are the details of the new contact:\n"+f"{new_contact}")
          i += 1
```

### Modify a contact

```python
# POINT 3 - MODIFY A CONTACT

from collections import OrderedDict

def modify_contact(contacts):
    chosen_mod = None
    selection = None
    keys_order = list(contacts.keys())  # maintains the original order of the contacts
    contacts = OrderedDict(contacts)

    while True:
        modify = input("Select here the Name or Surname of the contact you want to modify: ").strip()

        potential_matches = {}
        for key, value in contacts.items():
            name_surname = key.split()
            if (modify.lower() in name_surname[1].lower()) or (modify.lower() in name_surname[2].lower()):
                potential_matches[key] = value

        if len(potential_matches) == 0:
            print("No contact found with this name or surname.\n")
        else:
            break

    if len(potential_matches) == 1:
        for key in keys_order:
            if key in potential_matches:
                value = contacts[key]
                print("\n", key, value[0], value[1])

                print("\nHere is the contact you selected. What do you want to modify?")
                print("Please, digit one of the number of the following options:\n")

                options = [
                    'Name',
                    'Surname',
                    'Phone number',
                    'Email address',
                ]

                c = 1
                for x in options:
                    print(f"{c} " + x)
                    c += 1

                while chosen_mod is None:
                    try:
                        chosen_mod = int(input("\n"))
                        if chosen_mod == 1:
                            new_name = input("You decided to change the name of the contact. Please, digit here the new name: ")

                            split = key.split(sep=" ")
                            split[1] = new_name
                            unsplit = " ".join(split)
                            contacts[unsplit] = contacts.pop(key)
                            keys_order[keys_order.index(key)] = unsplit
                            contacts = OrderedDict((key, contacts[key]) for key in keys_order)
                            return contacts

                        if chosen_mod == 2:
                            new_surname = input("You decided to change the surname of the contact. Please, digit here the new surname: ")

                            split = key.split(sep=" ")
                            split[2] = new_surname
                            unsplit = " ".join(split)
                            contacts[unsplit] = contacts.pop(key)
                            keys_order[keys_order.index(key)] = unsplit
                            contacts = OrderedDict((key, contacts[key]) for key in keys_order)
                            return contacts


                        if chosen_mod == 3:
                            new_phone = input("You decided to change the phone number of the contact. Please, digit here the new phone number: ")

                            value[0] = new_phone
                            contacts[key] = value
                            return contacts


                        if chosen_mod == 4:
                            new_email = input("You decided to change the email address of the contact. Please, digit here the new email address: ")

                            value[1] = new_email
                            contacts[key] = value
                            return contacts


                    except ValueError:
                        print("Oops! Something went wrong!\nPlease, insert a number for the desired option!")
                        chosen_mod = None

    else:
        print("More than one contact has been found with this name or surname: ")
        for name, number in potential_matches.items():
            print(f"{name}: {number[0]} {number[1]}")

        specific_mod = None
        while specific_mod is None:
            try:
                specific_mod = input("Please, specify here the index number of the contact you want to modify: ").strip()
                found = False
                for key in potential_matches.keys():
                    if key.startswith(specific_mod + " "):
                        found = True
                        value = contacts[key]

                        print("\n", key, value[0], value[1])

                        print("\nHere is the contact you selected. What do you want to modify?")
                        print("Please, digit one of the number of the following options:\n")

                        options = [
                            'Name',
                            'Surname',
                            'Phone number',
                            'Email address',
                        ]

                        c = 1
                        for x in options:
                            print(f"{c} " + x)
                            c += 1

                        while chosen_mod is None:
                            try:
                                chosen_mod = int(input("\n"))
                                if chosen_mod == 1:
                                    new_name = input("You decided to change the name of the contact. Please, digit here the new name: ")

                                    split = key.split(sep=" ")
                                    split[1] = new_name
                                    unsplit = " ".join(split)
                                    contacts[unsplit] = contacts.pop(key)
                                    keys_order[keys_order.index(key)] = unsplit
                                    contacts = OrderedDict((key, contacts[key]) for key in keys_order)
                                    return contacts


                                if chosen_mod == 2:
                                    new_surname = input("You decided to change the surname of the contact. Please, digit here the new surname: ")

                                    split = key.split(sep=" ")
                                    split[2] = new_surname
                                    unsplit = " ".join(split)
                                    contacts[unsplit] = contacts.pop(key)
                                    keys_order[keys_order.index(key)] = unsplit
                                    contacts = OrderedDict((key, contacts[key]) for key in keys_order)
                                    return contacts


                                if chosen_mod == 3:
                                    new_phone = input("You decided to change the phone number of the contact. Please, digit here the new phone number: ")

                                    value[0] = new_phone
                                    contacts[key] = value
                                    return contacts


                                if chosen_mod == 4:
                                    new_email = input("You decided to change the email address of the contact. Please, digit here the new email address: ")

                                    value[1] = new_email
                                    contacts[key] = value
                                    return contacts


                            except ValueError:
                                print("Oops! Something went wrong!\nPlease, insert a number for the desired option!")
                                chosen_mod = None
                if not found:
                    print("Invalid index. Please insert a correct ID from the list.")
                    specific_mod = None

            except ValueError:
                print("Invalid input. Please specify a correct index number!")
                specific_mod = None
```

### Delete a contact

```python
# POINT 4 - DELETE A CONTACT

def delete_contact(contacts):

  del_dict = contacts.copy()
  print(f"{warning} ", "WARNING: Once the contact is deleted, it won't be possible to regain it! ", f"{warning}\n")

  delete = input("Please, insert here the name or surname of the contact you want to delete: ")


  potential_matches = {}
  for key, value in del_dict.items():
        name_surname = key.split()
        if delete.lower() in name_surname[1].lower() or delete.lower() in name_surname[2].lower():
          potential_matches[key] = value

  if len(potential_matches) == 0:
      print("No contact found with this name or surname.")
  elif len(potential_matches) == 1:
      contact_to_delete = list(potential_matches.keys())[0]
      del del_dict[contact_to_delete]
      print(f"Contact '{contact_to_delete}' has been deleted.")
      return del_dict
  else:
      print("More than one contact has been found with this name or surname:")
      for name, number in potential_matches.items():
          print(f"{name}: {number[0]} {number[1]}")
      specific_delete = int(input("Please, specify the index number of the contact you want to delete: "))
      for key in list(del_dict.keys()):
        if str(specific_delete) in key:
          del del_dict[key]
          print(f"Contact '{specific_delete}' has been deleted.")
          return del_dict
```

### Search for a contact using name or surname

```python
# POINT 5 - SEARCH FOR A CONTACT USING NAME OR SURNAME

def search_contact(contacts):

  search = input("Search for a specific contact, inserting here the name or surname of the contact! \n")

  for key, value in contacts.items():
    if search.lower() in key.lower(): # This allows the users to digit lower cases too.
        print(key, value[0], value[1])
```

### Save contacts in a JSON file
Since it is important to save the contacts to avoid losing them, I decided that the saving message will appear every time the programm ends

```python
# POINT 7 - SAVE CONTACTS IN A JSON FILE (IT WILL APPEAR EVERY TIME THE PROGRAMM ENDS)
# https://colab.research.google.com/notebooks/io.ipynb#scrollTo=p2E4EKhCWEC5

from google.colab import files
import json

def save_file(contacts):
    filename = "contacts.json"
    with open(filename, 'w') as json_file:
         json.dump(contacts, json_file, indent=6) #indent allows the correct formatting in a json file
    files.download(filename)
```

### Request for a new job

```python
# MORE JOB

def more_job(more, contacts):
    more = str(more)
    if more.lower() in ["yes", "y"]:
        return True  # continue the programm
    else:
        print("Remember to save your contacts!")
        save_file(contacts)
        print(f"\nThank you for using ContactManager! Have a nice day! {bye}")
        return False  #stop the programm
```
